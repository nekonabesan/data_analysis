### 9.1.3 バックテストを実行する

```python
name = "新日本科学"
code = "2395.T"

# チャート表示開始日
display_start_date = dt.datetime(2021, 1, 1)
# データ取得終了日(現在の日付を取得)
end_date = dt.datetime(2022, 3, 31)
# データ取得開始日
start_date = display_start_date - dt.timedelta(days=300)

### 新日本科学 (2395.T)
df = get_market_data.get_data_from_yfinance(code, start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))
if isinstance(df.columns, pd.MultiIndex):
    df = df.droplevel(level=1, axis=1)
close = df['Close']

data = df[display_start_date:end_date]

bt = Backtest(data, SmaCross, trade_on_close=True)

# バックテストを実行
result = bt.run()

# 結果を表示
print(result)

# バックテストの結果をプロット
bt.plot()
```

### 9.1.4 パラメータの最適化

```python
result = bt.optimize(ns=range(5, 15, 5),
                    nl=range(5, 75, 5),
                    maximize='Return [%]',
                    constraint=lambda r: r.ns < r.nl)
print(result)
bt.plot()
```
